```python
# sample 파일 총 24개를 읽어들여서
# sample_total 데이터프레임에 합쳐서 넣을 겁니다.
# 최종 컬럼 : 국적코드, 성별, 입국객수, 기준년월, 국적명
# sample_total 조회
# 기준년월 : 각 파일당 2019년 12월 ~ 2020년 12월
```

```python
# 국적정보 파일 불러와서 codemaster에 저장
codemaster = pd.read_excel('C:/DEV/STUDY/StudyPython22.06/day1/files2/sample_codemaster.xlsx')
codemaster
```

```python
# sample_total 빈 데이터 프레임 작성
sample_total = pd.DataFrame({'국적코드':[], '성별':[], '입국객수':[]})
sample_total
```

```python
# 파일경로 리스트 작성하여 files에 저장
files = []
for i in range(24):
    a = 'C:/DEV/STUDY/StudyPython22.06/day1/files2/sample('+str(i)+').xlsx'
    files.append(a)
files
```

```python
# 날짜 리스트 작성
date_ = []
year = [2019, 2020]
month = [i for i in range(1,13)]
for i in year:
    for j in month:
        a = str(i)+'-'+str(j).zfill(2)
        date_.append(a)

print(date_)
```

```python
# 파일 불러와서 sample_total에 통합
for i in range(24):
    sp_ = pd.read_excel(files[i],
                       header = 1,
                       skipfooter = 2,
                       usecols = 'A:C')
    
    sp_['기준년월'] = date_[i]
    
    sample_total = pd.concat([sample_total, sp_], ignore_index=True)

sample_total = pd.merge(sample_total, codemaster, how='left')

sample_total
```

```python
# 강사님 풀이
```

```python
code_master2 = pd.read_excel('C:/DEV/STUDY/StudyPython22.06/day1/files2/sample_codemaster.xlsx')
i = 0
sample_total = pd.DataFrame()

for yyyy in range(2019, 2021, 1):
    for mm in range(1, 13, 1):
        
        file_path = 'C:/DEV/STUDY/StudyPython22.06/day1/files2/sample({}).xlsx'.format(i)
        sample2 = pd.read_excel(file_path, header=1, skipfooter=2, usecols='A:C')
        i += 1
        
        sample2['기준년월'] = str(yyyy)+'-'+str(mm).zfill(2)
        sample_end = pd.merge(left = sample2,
                              right = code_master2,
                              how = 'left',
                              left_on = '국적코드',
                              right_on = '국적코드')
        
        sample_total = pd.concat([sample_total, sample_end], ignore_index=True)
        
print(len(sample_total))
sample_total
```

# 본인부담금 상위권 10개 진료과목

```python
file_path = 'C:/DEV/STUDY/StudyPython22.06/day7/files/df_20TABLE_kor_명세서일반내역.xlsx'
df_20_tbl = pd.read_excel(file_path)
df_20_tbl
```

```python
df_my_cost = df_20_tbl.copy()
```

```python
df_my_cost_mean = df_my_cost.groupby('진료과목코드')['심결본인부담금'].mean()
df_my_cost_mean.sort_values(ascending=False).head(10)
```

```python
# 본인부담금 상위 10개 진료과목코드 [2,3,4,6,7,8,24,50,54,55]
```

```python
hos_code_1 = [2,3,4,6,7,8,24,50,54,55]
```

```python
df_mycost_result = pd.DataFrame()
```

```python
for i in hos_code_1:
    df_temp = df_result[df_result['진료과목코드'] == i]
    df_mycost_result = pd.concat([df_mycost_result,df_temp], ignore_index=True)
```

```python
df_mycost_result.groupby('진료종류')['금액'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(16,10))
plt.rc('font', family='Malgun Gothic')
sns.barplot(x='진료종류',y='금액',data=df_mycost_result)
plt.title('본인부담금 상위 10개 진료과목에 대한 진료종류금액 분석')
```

```python
'''
본인부담금 상위 10개 진료과목에 대해서 입원료가 가장 본인 부담금액이 높고 
그 다음이 정신요법료인 것으로 나타남.
입원료가 가장 본인 부담금액이 높게 나타나는 것은 장기입원 환자의 경우 
입원료가 점점 늘어나기때문에 그런 것으로 보입니다.
입원료의 보험자 부담금 비율이 늘어나면 환자의 부담을 줄일 수 있을 것으로 생각됩니다.
'''
```

# 전체파일 통합

```python
df = pd.DataFrame()
```

```python
for i in range(80):
    file_path = 'C:/DEV/STUDY/StudyPython22.06/day8/data/trfcard({}).csv'.format(i)
    df_temp = pd.read_csv(file_path)
    df = pd.concat([df,df_temp], ignore_index=True)
```

# 컬럼명 한글로 변환

```python
file_path = 'C:/DEV/STUDY/StudyPython22.06/day8/data/데이터_정의서.xlsx'
col_kor_name = pd.read_excel(file_path,
                             header=11,
                             usecols='A,B')
col_kor_name
```

```python
col_kor_name_list = col_kor_name.to_dict('records')
col_kor_name_list
```

```python
# 깨끗한 딕셔너리 변수 생성
new_dict = {}
```

```python
for i in col_kor_name_list:
    value, key = i.values()
    new_dict[key] = value
    
df_kor = df.rename(columns=new_dict)
df_kor
```

```python
df_kor.to_csv('C:/DEV/STUDY/StudyPython22.06/day8/data/trfcard_result.csv', index=False)
```

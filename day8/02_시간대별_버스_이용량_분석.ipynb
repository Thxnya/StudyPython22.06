```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
# 파일 불러오기
file_path = 'C:/DEV/STUDY/StudyPython22.06/day8/data/trfcard_result.csv'
df_cp = pd.read_csv(file_path)
df_cp
```

```python
df = df_cp.copy()
df
```

```python
df.info()
```

```python
df['승차시각'] = df['승차시각'].astype(str)
df['승차시각']
```

```python
df['하차시각'] = df['하차시각'].astype(str)
df['하차시각']
```

```python
df['승차월'] = df['승차시각'].str[4:6]
df
```

```python
df['승차일'] = df['승차시각'].str[6:8]
df
```

```python
'''
연령 None :
에러가 발생하여 두번 찍은 경우 연령코드가 에러로 뜸(None)
이 경우 제거해도 분석에 지장이 없을 것으로 판단
결측처리 > 제거
'''

df = df[df['승객연령'] != 'None']
df
```

```python
df['승차시간'] = df['승차시각'].str[8:10]
df
```

```python
df['승차분'] = df['승차시각'].str[10:12]
df
```

# 기준월 및 기준일자별 버스 이용량분석 비교

```python
df_month_day = df.pivot_table(index='승차월',
                          columns='승차일',
                          values='노선명',
                          aggfunc='count')
df_month_day = df_month_day.fillna(0)
df_month_day
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_month_day, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('월별 및 일별 버스 이용량분석 비교')
plt.show()
```

1월과 2월 초~중순까지는 버스 이용객의 트렌드가 유사하지만
2월 말부터 코로나 확산의 영향으로 버스 이용객수가 줄어든 것으로 보입니다.

# 기준일 및 시간대별 버스 이용량 분석 비교

```python
df_day_hour = df.pivot_table(index='승차일',
                          columns='승차시간',
                          values='노선명',
                          aggfunc='count')
df_day_hour = df_day_hour.fillna(0)
df_day_hour
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_day_hour, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('기준일 및 시간대별 버스 이용량 분석 비교')
plt.show()
```

버스 승차일 기준으로 본 버스 이용량은 데이터에서 2월의 경우에 30, 31일이 없고,
3월의 경우에 14 ~ 24일의 데이터가 없기때문에 분석에 어려움이 있습니다.
시간대별 버스 이용량은 13 ~ 15시까지 점점 증가하다가 이후로 다시 감소하는 것으로 보입니다.

# 기준시간 및 시간(분)별 버스 이용량 분석 비교

```python
df_hour_min = df.pivot_table(index='승차시간',
                          columns='승차분',
                          values='노선명',
                          aggfunc='count')
df_hour_min = df_hour_min.fillna(0)
df_hour_min
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_hour_min, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('기준시간 및 시간(분)별 버스 이용량 분석 비교')
plt.show()
```

시간(분)별로 보는 버스 이용량은 시간대별로 유사하고,
시간으로 보는 버스 이용량의 경우에 앞에서 분석한 것과 마찬가지로
13 ~ 15시까지 점점 증가하다가 이후로 다시 감소하는 것으로 보입니다.
